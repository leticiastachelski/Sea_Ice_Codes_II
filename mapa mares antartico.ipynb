```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
fig=plt.figure(figsize=(5,10))
m= Basemap(projection='spstere',resolution='f',boundinglat=-50, round=True,lon_0=180)
#-----------------------------------------
m.drawlsmask()
#m.drawcoastlines()
m.drawparallels(np.arange(-80.,81.,20.),labels=[True,True,True,True])
m.drawmeridians(np.arange(0.,360.,20.),labels=[True,True,True,True])
#----------------------------------
x, y = m(-40, -68)
plt.text(x, y, 'WS', fontsize=11)
#----------------------------------
x1, y1 = m(-85, -65)
plt.text(x1, y1, 'BS', fontsize=11)
#----------------------------------
x2, y2 = m(-115, -65)
plt.text(x2, y2, 'AS', fontsize=11)
#----------------------------------
x2, y2 = m(-162, -66)
plt.text(x2, y2, 'RS', fontsize=11)
#----------------------------------
x3, y3 = m(160, -64)
plt.text(x3, y3, 'WPO', fontsize=11)
#----------------------------------
x4, y4 = m(50, -65)
plt.text(x4, y4, 'IO', fontsize=11)
#----------------------------------

plt.show()
```
